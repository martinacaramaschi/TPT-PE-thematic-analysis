# PE and TPT cleaning

## **Removing specific series using a keyword in the title**

```python
# PE
import pandas as pd

# List of keywords or phrases you want to remove from titles
keywords_to_remove1 = ['News', 'SIGNING OFF', 'PEOPLE', 'People']
keywords_to_remove2 = ['Reply to comment', 'Erratum', 'book review']



# Create a condition that checks if any of the keywords are in the 'title' column
condition1 = pe_data['title'].str.contains('|'.join(keywords_to_remove1), case=True, na=False)
condition2 = pe_data['title'].str.contains('|'.join(keywords_to_remove2), case=False, na=False)

# Filter the dataset by inverting the condition to keep rows that don't match the keywords
pe_data_cleaned = pe_data[~condition1].copy()
pe_data_cleaned = pe_data_cleaned[~condition2]

# Display the cleaned dataset
print(pe_data_cleaned.shape)
```

```python
# TPT
import pandas as pd

# List of keywords or phrases you want to remove from titles
keywords_to_remove1 = ['Correction', 'BOOKS FOR YOUNG PEOPLE']
keywords_to_remove2 = ['century textbook illustrations']



# Create a condition that checks if any of the keywords are in the 'title' column
condition1 = tpt_data['title'].str.contains('|'.join(keywords_to_remove1), case=True, na=False)
condition2 = tpt_data['title'].str.contains('|'.join(keywords_to_remove2), case=False, na=False)

# Filter the dataset by inverting the condition to keep rows that don't match the keywords
tpt_data_cleaned = tpt_data[~condition1].copy()
tpt_data_cleaned = tpt_data_cleaned[~condition2]

# Display the cleaned dataset
print(tpt_data_cleaned.shape)
```

## **Removing raws without any author**

```python
# Remove rows where 'authors' is an empty string
# PE
pe_data_cleaned = pe_data_cleaned.dropna(subset=['authors'])
pe_data_cleaned = pe_data_cleaned[pe_data_cleaned['authors'] != '']

# TPT
tpt_data_cleaned = tpt_data_cleaned.dropna(subset=['author_list'])
tpt_data_cleaned = tpt_data_cleaned[tpt_data_cleaned['author_list'] != '']
```

```python
print("PE: ", pe_data_cleaned.shape)
print("TPT: " ,tpt_data_cleaned.shape)
```

```python
# Remove rows where 'fulltext' is an empty string
# PE
pe_data_cleaned = pe_data_cleaned.dropna(subset=['fulltext'])
pe_data_cleaned = pe_data_cleaned[pe_data_cleaned['fulltext'] != '']

# TPT
tpt_data_cleaned = tpt_data_cleaned.dropna(subset=['fulltext'])
tpt_data_cleaned = tpt_data_cleaned[tpt_data_cleaned['fulltext'] != '']

print("PE: ", pe_data_cleaned.shape)
print("TPT: " ,tpt_data_cleaned.shape)
```

## **counting words number**

we counted how many rows will be eliminated by setting different thresholds: 300-400-500

```python
# identifying rows having fulltext less than "number" words

# TPT
# Assuming tpt_data_cleaned is your DataFrame and 'fulltext' contains the text
tpt_data_cleaned['word_count'] = tpt_data_cleaned['fulltext'].apply(lambda x: len(str(x).split()))

# Filter rows where word count is less than [number]
rows_less_than_number = tpt_data_cleaned[tpt_data_cleaned['word_count'] < 300]
rows_less_than_number.shape

# results: 500 --> 55 / 400 --> 29 / 300 --> 17
```

```python
# PE
# Assuming tpt_data_cleaned is your DataFrame and 'fulltext' contains the text
pe_data_cleaned['word_count'] = pe_data_cleaned['fulltext'].apply(lambda x: len(str(x).split()))

# Filter rows where word count is less than [number]
rows_less_than_number = pe_data_cleaned[pe_data_cleaned['word_count'] < 300]

rows_less_than_number.shape

# results: 500 --> 252 / 400 --> 153 / 300 --> 76
```

```python
# Remove rows from pe_data_cleaned and tpt_data_cleanedwhere 'fulltext' is less than 500 words

#PE
pe_data_cleaned = pe_data_cleaned[pe_data_cleaned['word_count'] >= 500]
print("PE: ", pe_data_cleaned.shape)

# TPT
tpt_data_cleaned = tpt_data_cleaned[tpt_data_cleaned['word_count'] >= 500]
print("TPT: ", tpt_data_cleaned.shape)
```
